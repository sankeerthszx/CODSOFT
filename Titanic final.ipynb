```python
num_df = df[["PassengerId","Age","Fare","Ticket"]]
cat_df = df[["Sex","Pclass","Cabin","Embarked","SibSp","Parch"]]
print(num_df.head())
print(cat_df.head(10))
```

```python
#removing null values in cat_df
df['Cabin']=cat_df['Cabin'].fillna(cat_df['Cabin'].mode()[0])
df['Embarked']=cat_df['Embarked'].fillna(cat_df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
#removed the null values of age by mean of age
df["Age"]=num_df["Age"].fillna(num_df["Age"].mean())
df.isna().sum()
```

```python
#feature scaleing using standardisation
#numeric data are standardised
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()


df['PassengerId'] = ss.fit_transform(df[['PassengerId']])
df['Age'] = ss.fit_transform(df[['Age']])
df['Fare'] = ss.fit_transform(df[['Fare']])
df.head()
```

```python
#one hot encoding for category
#convert catogorical data to numeric
df = pd.get_dummies(df,columns=['Sex', 'Pclass', 'Embarked', 'SibSp', 'Parch'],drop_first=True)
df.head()
```

```python
df=df.drop(['Name',"Ticket","Cabin","PassengerId"],axis = 1)


```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop("Survived",axis = 1)
y = df["Survived"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
survuval_pred = model.predict(x_test)
accuracy = accuracy_score(survuval_pred,y_test)
print("Accuracy of the model :",accuracy)
```
